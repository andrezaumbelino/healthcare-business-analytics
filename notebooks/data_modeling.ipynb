## Limpeza de Dados


### Strings

```python
df["Name"] = df["Name"].str.title()
df["Medical Condition"] = df["Medical Condition"].str.strip().str.title()
df["Doctor"] = df["Doctor"].str.strip().str.title()
df["Hospital"] = df["Hospital"].str.strip().str.title()
```

### Dates

```python
df["Date of Admission"] = pd.to_datetime(df["Date of Admission"], errors="coerce")
df["Discharge Date"] = pd.to_datetime(df["Discharge Date"], errors="coerce")
```

```python
df["length_of_stay"] = (df["Discharge Date"] - df["Date of Admission"]).dt.days
```

### Duplicadas

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
```

```python
df.duplicated(subset=[
    "Name",
    "Date of Admission",
    "Hospital",
    "Doctor"
]).sum()
```

```python
df[
    df.duplicated(
        subset=["Name", "Date of Admission", "Hospital", "Doctor"],
        keep=False
    )
].sort_values(by=["Name", "Date of Admission"]).head(20)
```

```python
df_check = df.groupby(
    ["Name", "Date of Admission", "Hospital"]
)["Age"].nunique().reset_index()

df_check[df_check["Age"] > 1].nunique()
```

```python
df = (
    df.sort_values(by="Age")
    .drop_duplicates(
        subset=["Name", "Date of Admission", "Hospital", "Doctor"],
        keep="first"
    )
)
```

## Métricas Adicionais

### Duração Internamento 

```python
df["length_of_stay"] = (
    df["Discharge Date"] - df["Date of Admission"]
).dt.days
```

```python
(df["length_of_stay"] == 0).sum()
```

### Custo por Dia

```python
df["cost_per_day"] = (
    df["Billing Amount"] / df["length_of_stay"]
) 
```

## Estruturação Modelo

### Dimensões

```python
d_paciente = df[[
    "Name",
    "Age",
    "Gender",
    "Blood Type"
]]

d_paciente["patient_id"] = range(1, len(d_paciente) + 1)
```

```python
d_hospital = df[["Hospital"]].drop_duplicates().reset_index(drop=True)

d_hospital["hospital_id"] = range(1, len(d_hospital) + 1)
```

```python
d_doctor = df[["Doctor"]].drop_duplicates().reset_index(drop=True)

d_doctor["doctor_id"] = range(1, len(d_doctor) + 1)
```

```python
d_condition = df[["Medical Condition"]].drop_duplicates().reset_index(drop=True)

d_condition["condition_id"] = range(1, len(d_condition) + 1)
```

```python
d_admission = df[["Admission Type"]].drop_duplicates().reset_index(drop=True)

d_admission["admission_type_id"] = range(1, len(d_admission) + 1)
```

```python
d_insurance = df[["Insurance Provider"]].drop_duplicates().reset_index(drop=True)

d_insurance["insurance_id"] = range(1, len(d_insurance) + 1)
```

### Fato

```python
f_internacoes = df.copy()

f_internacoes = f_internacoes.merge(d_paciente, on=["Name", "Age", "Gender", "Blood Type"], how="left")
f_internacoes = f_internacoes.merge(d_hospital, on="Hospital", how="left")
f_internacoes = f_internacoes.merge(d_doctor, on="Doctor", how="left")
f_internacoes = f_internacoes.merge(d_condition, on="Medical Condition", how="left")
f_internacoes = f_internacoes.merge(d_admission, on="Admission Type", how="left")
f_internacoes = f_internacoes.merge(d_insurance, on="Insurance Provider", how="left")
```

```python
f_internacoes["internacao_id"] = range(1, len(f_internacoes) + 1)
```

```python
f_internacoes.columns.tolist()
```

```python
f_internacoes = f_internacoes.rename(columns={
    "Billing Amount": "billing_amount",
    "Room Number": "room_number",
    "Medication": "medication",
    "Test Results": "test_result"
})
```

```python
f_internacoes = f_internacoes[[
    "internacao_id",
    "patient_id",
    "hospital_id",
    "doctor_id",
    "condition_id",
    "admission_type_id",
    "insurance_id",
    "admission_date",
    "discharge_date",
    "billing_amount",
    "room_number",
    "length_of_stay",
    "cost_per_day",
    "medication",
    "test_result"
]]
```

## Exportar tabelas

```python
f_internacoes.to_csv("../data/f_internacoes.csv", index=False)
d_paciente.to_csv("../data/d_paciente.csv", index=False)
d_hospital.to_csv("../data/d_hospital.csv", index=False)
d_doctor.to_csv("../data/d_doctor.csv", index=False)
d_condition.to_csv("../data/d_condition.csv", index=False)
d_admission.to_csv("../data/d_admission_type.csv", index=False)
d_insurance.to_csv("../data/d_insurance.csv", index=False)
```
